```python
import sys
sys.path.insert(0,'..')
import gymnasium as gym
import numpy as np
from src.dont_starve_2D import DontStarveSingleAgentEnv

# Create the environment
env = DontStarveSingleAgentEnv(grid_size=10, max_days=100, render_mode="ansi")

# Reset the environment
obs, info = env.reset(seed=42)

# Run some random actions
terminated = truncated = False
total_reward = 0

while not (terminated or truncated):
    # Sample a random action
    action = env.action_space.sample()
    
    # Take a step in the environment
    obs, reward, terminated, truncated, info = env.step(action)
    
    # Update total reward
    total_reward += reward
    
    # Print information
    print(f"Action: {action}, Reward: {reward:.2f}, Total: {total_reward:.2f}")
    print(f"HP: {env.hp}, Satiety: {env.satiety}, Hydration: {env.hydration}, Day: {env.days}")
    print(f"Inventory: Food: {env.inventory['food']}, Water: {env.inventory['water']}")
    
    # Render the environment (if render_mode is set)
    render_result = env.render()
    if render_result:
        print(render_result)
    
    print("-" * 30)

print(f"Episode finished! Total reward: {total_reward:.2f}")
```
